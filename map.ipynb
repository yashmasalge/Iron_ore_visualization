```python
ds = pd.read_csv('Iron_Ore_1.csv')
ds.head()
```

```python
latitude = ds['LATDD'].mean()
longitude = ds['LONDD'].mean()
my_map1 = folium.Map(location = [latitude, longitude],
                                        zoom_start = 6 )
display(my_map1)
```

```python
folium.CircleMarker(location = [latitude, longitude],
                    radius = 50, popup = ' FRI ')
display(my_map1)
```

```python
my_map2 = folium.Map(location = [latitude, longitude],
                                         zoom_start = 6)
 
# CircleMarker with radius
folium.Marker(location = [latitude, longitude],
                     popup = ' Yash MAsalge ').add_to(my_map2)
display(my_map2)
```

```python
locations = ds[['LATDD','LONDD']]
locationlist = locations.values.tolist()
len(locationlist)
locationlist[6]
```

```python
map = folium.Map(location=[latitude, longitude], zoom_start=6)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=ds['STATE'][point]).add_to(map)
    display(map)
```

```python
for lat, lon, locality, state, metal in zip(ds['LATITUDE'], ds['LONGITUDE'], ds['LOCALITY'], ds['STATE'], ds['METALLOGENESIS']):
    folium.CircleMarker(
        [latitude, longitude],
        popup = ('STATE: ' + str(state).capitalize() + '<br>'
                 'Locality: ' + str(locality) + '<br>'
                 'metal: ' + str(metal) +'%'
                ),
        color='b',
        key_on = state,
        threshold_scale=[0,1,2,3],
        
        fill=True,
        fill_opacity=0.7
        ).add_to(map)

display(map)
```
